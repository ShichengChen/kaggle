# Titanic Feature Correlations Using Speedml

This notebook implements demos for the [Speedml](https://speedml.com) Python package for speed starting machine learning projects.

**Step 1.** Just import the [Speedml Python package](https://pypi.python.org/pypi/speedml). Add the Python notebook magic command for inline plots.

```python
from speedml import Speedml
```

**Step 2.** Prerequisites include loading the datasets. Ensuring that there are a few numerical features.

```python
sml = Speedml('../input/titanic/train.csv', 
              '../input/titanic/test.csv', 
              target = 'Survived', uid = 'PassengerId')
sml.shape()
```

**Step 3.** Call the correlate method.

```python
sml.plot.correlate()
```
